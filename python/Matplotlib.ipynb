```python
fig = plt.figure()

ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
y2= np.sin(x-1)
z2= np.cos(x-1)

# line style変更
ax1.plot(x,y,color="red")
ax1.plot(x,y2,color="c") # short color code (rgb cmyk)
ax2.plot(x,z,color="0.75") # gray scale between 0 and 1
ax2.plot(x,z2,color=(0,0.6,0.3)) # RGB　tuple between 0 and 1
```

```python
from matplotlib.colors import LinearSegmentedColormap

def generate_cmap(colors):
    """自分で定義したカラーマップを返す"""
    values = range(len(colors))

    vmax = np.ceil(np.max(values))
    color_list = []
    for v, c in zip(values, colors):
        color_list.append( ( v/ vmax, c) )
    return LinearSegmentedColormap.from_list('custom_cmap', color_list)

# 自分でカラーを設定
rkb = generate_cmap(["red","black","blue"])
```

```python
# plot totally 
sig_df_list = [immuno_sig_df,kegg_sig_df,go_sig_df,biocarta_sig_df, df_normalized ]

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(len(sig_df_list)+1 ,1, 1) 
ax = sns.barplot(data.index, np.log2(muse),color="blue")# width=1 棒の間を埋める
ax.set_xticks([])
ax.set_xlabel("log2(muse counts)")

for i in range(1, len(sig_df_list)+1):
    ax = fig.add_subplot(len(sig_df_list)+1 ,1,  i+1) 
    ax = sns.heatmap(sig_df_list[i-1].T, cmap=rkb, cbar=False,vmin=-3,vmax=3)
    ax.set_xticks([])
    
# save as png
plt.savefig('TMB_Immuno_score.png')
```
